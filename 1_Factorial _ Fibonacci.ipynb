#Factorial 

##Using For Loop

```
import time
```

```
def factorial_forloop1(n):
  fact = 1
  for i in range(1,n+1):
    fact *= i
  return fact  

factorial_forloop1(10)
```

```
tik = time.time()
factorial_forloop1(900)
tok = time.time()
print(tok-tik,"seconds")
```

```
def factorial_forloop2(n):
  f = 1
  print("Find factorial of number, n = ",n)
  print("initialized an integer variable, f = ",f)
  print("++++++Start for loop++++++")
  for i in range(1,n+1):

    f = f * i
    print("After ",i," run:","f = ",f)
  
  print("Factorial of ",n," is ",f)
```

```
tik = time.time()
factorial_forloop2(10)
tok = time.time()
print(tok-tik,"seconds")
```

##Using Recursion

```
def factorial_recursion(n):
  if(n==0 or n==1):
    return 1
  else:
    return n * factorial_recursion(n-1)
```

```
tik = time.time()
print(factorial_recursion(900))
tok = time.time()
print(tok-tik, "seconds")
```

```

def factorial_recursion_memoised(n):
  memo = {}
  memo[0] = 1
  memo[1] = 1
  if (n in memo):
    return 1

  else:
    memo[n] = n * factorial_recursion_memoised(n-1)
```

```
factorial_recursion_memoised(10)
```

##Using Recursion

```
# Function for nth Fibonacci number 
def Fibonacci(n):

	if n<0: 
		print("Incorrect input") 
	elif n==0: 
		return 0
	elif n==1: 
		return 1
	else:
		k = Fibonacci(n-1)+Fibonacci(n-2)
		return k

print(Fibonacci(9)) 
```
