```
import numpy as np
import pandas as pd
import os
import json
from pathlib import Path
import matplotlib.pyplot as plt
from matplotlib import colors
from IPython.display import display
from tqdm import tqdm_notebook
```

```
X_train = []
Y_train = []
X_test =  []
Y_test =  []
file =-1
```

```
data_path = Path('/content/drive/My Drive/ARC/data')
training_path = data_path / 'training'
evaluation_path = data_path / 'evaluation'
list_training_tasks = sorted(os.listdir(training_path))
list_evaluation_tasks = sorted(os.listdir(evaluation_path))

for i in tqdm_notebook(range(len(list_training_tasks)),total = len(list_training_tasks),unit = "files"):
  task_file = str(training_path / list_training_tasks[i])

  with open(task_file, 'r') as f:
    task = json.load(f)
    n_train_pairs = len(task['train'])
    n_test_pairs = len(task['test'])
    train = pd.DataFrame(task["train"])
    test = pd.DataFrame(task["test"])
    x_output = np.zeros(len(train))
    file =file+1;
    for i in range(len(train)):
      xinput = (train["input"])
      xoutput = (train["output"])

      xinput[i] = np.array(xinput[i])
      xoutput[i] =np.array(xoutput[i]) 

      xinput[i]  = np.append( xinput[i],xinput[i].shape)
      xinput[i]  = np.append( xinput[i],file)
      xoutput[i] = np.append( xoutput[i],xoutput[i].shape)
      xoutput[i] = np.append( xoutput[i],file)
      
      xinput[i]  = np.flip( xinput[i].flatten() )
      xoutput[i] = np.flip( xoutput[i].flatten() )
      
      X_train.append(xinput[i])  
      Y_train.append(xoutput[i])
    
    for i in range(len(test)):
      xinput  = (test["input"])
      xoutput = (test["output"])

      xinput[i] = np.array(xinput[i])
      xoutput[i] =np.array(xoutput[i]) 

      xinput[i]  = np.append( xinput[i],xinput[i].shape)
      xinput[i]  = np.append( xinput[i],file)
      xoutput[i] = np.append( xoutput[i],xoutput[i].shape)
      xoutput[i] = np.append( xoutput[i],file)
      
      xinput[i]  = np.flip(xinput[i].flatten() )
      xoutput[i] = np.flip( xoutput[i].flatten() )
      
      X_test.append(xinput[i])  
      Y_test.append(xoutput[i])
```

```
def newdata(X):
  f = X[0]
  n = X[1]
  m = X[2]
  data = X[3:]

  zero    = np.zeros(len(data),dtype=int)
  one     = np.zeros(len(data),dtype=int)
  two     = np.zeros(len(data),dtype=int)
  three   = np.zeros(len(data),dtype=int)
  four    = np.zeros(len(data),dtype=int)
  five    = np.zeros(len(data),dtype=int)
  six     = np.zeros(len(data),dtype=int)
  seven   = np.zeros(len(data),dtype=int)
  eight   = np.zeros(len(data),dtype=int)
  nine    = np.zeros(len(data),dtype=int)

  for i in range(len(data)):
    if(data[i]==0):
      zero[i] = 1
    if(data[i]==1):
      one[i] = 1
    if(data[i]==2):
      two[i] = 1
    if(data[i]==3):
      three[i] = 1
    if(data[i]==4):
      four[i] = 1
    if(data[i]==5):
      five[i] = 1
    if(data[i]==6):
      six[i] = 1
    if(data[i]==7):
      seven[i] = 1
    if(data[i]==8):
      eight[i] = 1
    if(data[i]==9):    
      nine[i] = 1

  return np.concatenate((zero,one,two,three,four,five,six,seven,eight,nine))
```

```
X_train_new = []
Y_train_new = []
for i in tqdm_notebook(range(len(X_train)),total = len(X_train),unit = "rows"):
  X_train_new.append(newdata(X_train[i]))
  Y_train_new.append(newdata(Y_train[i]))

X_test_new = []
Y_test_new = []
for i in  tqdm_notebook(range(len(X_test)),total = len(X_test),unit = "rows"):
  X_test_new.append(newdata(X_test[i]))
  Y_test_new.append(newdata(Y_test[i]))  
```

```
train_size_x = []
train_size_y = []
test_size_x = []
test_size_y = []

for i in range(len(X_train_new)):
  train_size_x.append(len(X_train_new[i]))
  train_size_y.append(len(Y_train_new[i]))

for i in range(len(X_test_new)):
  test_size_x.append(len(X_test_new))
  test_size_x.append(len(Y_test_new[i]))
```

```
import tensorflow as tf
```

```
input_train =  tf.convert_to_tensor(np.array(X_train_new[:5]))
output_train = tf.convert_to_tensor(np.array(Y_train_new[:5]))
input_test =   tf.convert_to_tensor(np.array(X_test_new[0]).reshape(1,90))
output_test =  tf.convert_to_tensor(np.array(Y_test_new[0]).reshape(1,810))
```

```
train_univariate = tf.data.Dataset.from_tensor_slices((input_train, output_train))
train = train_univariate.cache().shuffle(1000).batch(5).repeat()

val_univariate = tf.data.Dataset.from_tensor_slices((input_test, output_test))
val = val_univariate.batch(5).repeat()
```

```
model = tf.keras.models.Sequential(
                                    [tf.keras.layers.Dense(10, input_shape=[90]),
                                    tf.keras.layers.Dense(810)])

model.compile(optimizer='adam', loss='mae')
```

```
model.fit(train_univariate, epochs=100,
                      steps_per_epoch=100,
                      validation_data=val_univariate, validation_steps=50)
```
